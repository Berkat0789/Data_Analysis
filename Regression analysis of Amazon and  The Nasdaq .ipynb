Perform a regression analysis between Amazon and the Nasdaq

```python
import pandas as pd 
import numpy as np 
from pandas_datareader import data as web
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
tickers = ["AMZN", "^IXIC"]
dataframe = pd.DataFrame()
for t in tickers:
    dataframe[t] = web.DataReader(t, data_source = "yahoo", start = "2001-1-1")["Adj Close"]
```

```python
dataframe.info()
```

```python
dataframe
```

```python
#Normailize 
plt.figure(figsize=(16,8))
plt.plot(dataframe.index[0:], dataframe["^IXIC"], label = "NASDAQ")
plt.plot(dataframe.index[0:], dataframe["AMZN"], label = "AMAZON")
plt.xlabel("Time Line")
plt.ylabel("value")
plt.legend()
plt.show()
```
