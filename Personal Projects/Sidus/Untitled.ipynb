### The python jupyter notebook is to compare securties behavor and generate correlation analysis between them. the list of methods to do this is as follows

```python
import pandas as pd 
import numpy as np
from pandas_datareader import data as web
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
class Fund():
    """
    Must enter the securties list(Tickers), starting date(start), endate(end)
    """
    def __init__(self, tickers, start, stop):
        self.tickers = tickers
        self.start = start
        self.stop = stop
    def adj_Close(self):
        data_frame = pd.DataFrame()
        for t in self.tickers:
            data_frame[t] = web.DataReader(t, "yahoo", self.start, self.stop)["Adj Close"]
        return data_frame.plot(figsize = (16,6));
    def daily_returns(self):
        daily_returns = pd.DataFrame()
        returnsFrame = pd.DataFrame()
        for t in self.tickers:
            daily_returns[t] = web.DataReader(t, "yahoo", self.start, self.stop)["Adj Close"]
            returnsFrame[t] = daily_returns[t].pct_change()
        return returnsFrame.plot(kind = "kde", figsize = (16,8))
            
        
```

```python
Sidus = Fund(["IYR", "TLT", "HYG", "XLK", "LQD"], start = "2001-1-1", stop= "2019-1-1")
```
