```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
%matplotlib inline
```

```python
tips.head()
```

```python
sns.lmplot(x = "total_bill", y = "tip", data=tips, col = "time", hue = "sex", scatter_kws= {"s": 50})
```

### Add a grid instead of a single plot

```python
sns.lmplot(x = "total_bill", y = "tip", data=tips, col = "sex", row = "time")
```
