```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

# Functional Way

```python
plt.plot(x ,y, c = "r", ls = "--")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("title")
```

### Multiplot on the same canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y, c = "r")
plt.subplot(1,2,2)
plt.plot(y,x, c = "y")
```

## OPP for plotting Legengs and moving legends

```python
fig = plt.figure(figsize = (14,6))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel("X label")
axes.set_ylabel("Ylabel")
axes.set_title("Title here")
axes.plot(x,y, label = "Label 1")
axes.plot(y,x, label = "Label 2")
## loc number takes a few numbers for location 
plt.legend(loc = 4)
```

### Adding multiple axes to one canvas

```python
fig = plt.figure(figsize=(14,6))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes2 = fig.add_axes([0.15,0.25,0.3,0.6])
axes2.plot(y,x, c = "y")
```

# Subplot(S)

```python
fig, axes = plt.subplots(1,2, figsize = (16,8))
axes[0].plot(x,y)
axes[1].plot(y,x, c = "r")
plt.tight_layout()
```

# Plot apperance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, c= "orange", lw = 4, ls = "--", marker = "o", ms = 10, alpha = 0.4,
         mfc="r", mew = 5, mec = "g")
```

## X and y limits

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(y,x)
axes.set_xlim([10,20])
axes.set_ylim([2,5])
```
