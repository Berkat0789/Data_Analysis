```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

## set particular styles (sns.set_style)
#### darkgrid, whitegrid, ticks
### remove spines 

```python
plt.figure(figsize = (16,6))
sns.set_style("darkgrid")
sns.countplot(x = "sex", data=tips)
```

### Scale and context -> overide parameters and default
### context -> paper, Poster, fontsize

```python
sns.set_context(context="poster", font_scale= 0.6)
sns.countplot(x = "sex", data=tips)
```

```python
plt.figure(figsize=(15,6))
sns.set_context(context="poster", font_scale= 0.6)
sns.countplot(x = "sex", data=tips)
```

### Paletes and colors -> check matplot color map documentations 

```python
plt.figure(figsize=(16,6))
sns.lmplot(x = "total_bill", y = "tip", data=tips, aspect= 0.9, size=8, hue = "sex",scatter_kws={"s": 50}, palette= "seismic")
```
