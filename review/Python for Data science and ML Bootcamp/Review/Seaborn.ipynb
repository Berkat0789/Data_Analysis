```python
import seaborn as sns
%matplotlib inline
```

## Distrobution Plots 

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

### Dis Plot -> Univariate (One variable /column form data set -> provides a kDE)
### Histogram with a KDE

```python
sns.distplot(tips["total_bill"])
## line is a KDE -> Kernal Density distribution 
```

#  Joint Plot -> cobines two different distibution plots -> bit variate-2 vaiables
## Plot these the same as a scatter -> data, x and y column names

### uSe the kind parameter to create a plot of a different type-> default is a scatter can be hex 

```python
sns.jointplot(x = "total_bill", y ="tip", data= tips, height=8)
```

##### Differenct Kind

```python
sns.jointplot(x = "total_bill", y ="tip", kind = "hex",data= tips, height=8);
```

```python
sns.jointplot(x = "total_bill", y ="tip",data= tips, kind="kde" ,height=8)
```

```python
sns.jointplot(x = "total_bill", y ="tip",data= tips, kind="reg" ,height=8)
```

# Pair plots -> pair wise relations scess a data frame -> joint plot for every combo in the frame
###  "HUE" USE A CATEGORICAL COLUMN TO ADJUST THE DATA, FOR EXAMPLE SEX MALE OR FEMALE

```python
sns.pairplot(tips, hue="sex")
```

# Categorical Date Plotting  -> distributon based on categories
#### cinsideraes the averaege of a category as below SEX and the total bill  
#### Estimator that is aggregate of the data -. default of is mean -> can pass in your own funciton or one form np
#### same as setting up the join plot where x is a category column and y is a numerical 

```python
import numpy as np
```

```python
sns.barplot(x = "sex", y = "total_bill", data=tips, estimator= np.mean)
```

## count plot

## -> plots the count of the paticular category in a data set for example Male vs female -> how many males vs females

### only need the x axis -> the count will be the y  axis

```python
sns.countplot(x = "sex", data=tips)
```

# Box plots

### shoes the curtails of the data set and the wiskers extend to show the rest the others are the outlyers 
## X Y and data 

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, hue ="smoker")
```

# Violen Plot -> shows the distributon on the points on its side
### HUE with a cateogory data
### Split to diaply the distribution on each violen

```python
sns.violinplot(x ="day", y = "total_bill", data = tips, hue="sex", split = True )
```

# Strip plot - >displays a chart that wull show point in a line but may bee to cluttered if there is al arge data set so stay away if so. use hue, also jitter (disperse points to show more)

takes in category, and a data set(number) ->hue with a catefory

```python
sns.stripplot(x = 'day', y = "total_bill", data=tips, hue = "sex", jitter=True)
```

# swarmPlot -> same as strip plot but tries to fit the data in a horizontal linea bit batter

```python
sns.swarmplot(x = 'day', y = "total_bill", data=tips, hue = "sex")
```

# catPlot( formally Factor plot)  -> used to call other plots bar, violin, strip etc

```python
sns.catplot(x = 'day', y = "total_bill", kind = "swarm",data = tips, hue = "sex")
```
