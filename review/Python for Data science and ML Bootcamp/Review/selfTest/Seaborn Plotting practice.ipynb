## Plot the following types of graph with the tips data
## Distribution, Joint, Pair, bar, count, box,violen,strip, swarm,catplot

## Afte the above we will go into matrix plotting and wil create a - heatmap, piviottable, and a cluster map

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

## distrubution

```python
plt.figure(figsize = (16,6))
sns.distplot(tips["total_bill"], kde=False, bins = 20, color= "r")
```

# Joint

```python
sns.jointplot(x = "total_bill", y = "tip", data=tips, kind="reg")
```

# Pair 

```python
sns.pairplot(tips, hue = "sex", palette="rainbow")
```

# bar 

```python
sns.barplot(x = "sex", y = "total_bill", data=tips, palette="coolwarm", hue="smoker")
```

# box

```python
sns.boxplot(x = "sex", y = "tip", data=tips)
```

# count

```python
sns.countplot(x = "sex", data=tips, palette="rainbow", hue="smoker")
```

## Violen

```python
sns.violinplot(x = "sex", y = "total_bill", data=tips, hue = "smoker", split=True)
```

## Strip

```python
sns.stripplot(x = "sex", y = "total_bill", data=tips, jitter=True)
```

## Swarm

```python
sns.swarmplot(x = "sex", y = "tip", data=tips)
```

## CatPlot

```python
sns.catplot(x = "sex", y = "tip", data=tips, kind="swarm")
```

# Using martixes plots for visualization -> using the flights data

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

# heatmap -> use the correlation of the data set to plot

```python
sns.heatmap(flights.corr(), cmap="coolwarm", annot=True)
```

# HeatMap Piviot table

```python
flights_piv = flights.pivot_table(index= "month", columns= "year", values="passengers")
```

```python
sns.heatmap(flights_piv, cmap="rainbow")
```

## Cluster Map with piviot Table

```python
sns.clustermap(flights_piv, cmap = "coolwarm", standard_scale=1)
```

## Regression Plots

```python
sns.lmplot(x = "total_bill", y = "tip", data=tips, hue="sex", col="smoker", palette="seismic", scatter_kws={"s":40}, aspect=0.9, size=6)
```
