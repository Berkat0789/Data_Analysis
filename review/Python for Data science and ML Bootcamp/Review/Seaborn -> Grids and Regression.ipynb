```python
import seaborn as sns 
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
%matplotlib inline
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

## PairGrid -> same as pair plot but cab be customized with other plot types

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
# scatter on the upper middle and lower
```

## what is you want to specify whats on the upper middle and lower 

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

# Facit Grid -> breaks a graph down into category  as well as plots the  the plot type

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
gri = sns.FacetGrid(tips, col="time", row = "smoker")
gri.map(sns.distplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
```

## Add Hue and split using the columns col ="sex"
