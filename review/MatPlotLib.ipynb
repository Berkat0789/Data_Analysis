```python
import matplotlib.pyplot as plt
%matplotlib inline
#allows you to see your plots directly in the notebook
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Testing mat plot\n Yes we are", color = "r")
plt.show()
```

```python
#multilot on the same canvas 
#sulot arguments -> Rows, Columns, plot number
plt.subplot(1,2,1)
plt.plot(x,y, color = "b")
plt.x_label("X label")
plt.subplot(1,2,2)
plt.plot(y,x, color = "r")
```

```python
#Object oriented method -> figure object is a blank objext
fig = plt.figure()
#fig add axes function -> takesin left, bottom, width, and height list
axes = fig.add_axes([0.5,0.5,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("Xlabel", color = "r")
axes.set_ylabel("Ylabel")
axes.set_title("Title")
```

```python
#creating multiplot with OOP
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.85,0.85])
axes.set_xlabel("X label")
axes.set_ylabel("y label")
axes.set_title("Main title")
axes.plot(x,y)
axes2 = fig.add_axes([0.2,0.45,0.43,0.42])
axes2.set_xlabel("X label")
axes2.set_ylabel("y label")
axes2.set_title("Main title")
axes2.plot(y,x)
plt.show()
```

```python
#create subplots using oop -> axes here is added automarically be the subplots method
fig,ax = plt.subplots(1,2)
ax[0].set_title("First")
ax[1].set_title("Second")
ax[0].plot(x,y, label = "y dot x")
ax[0].plot(y,x, label = "x dot y")
ax[1].plot(y,x)

plt.legend()

plt.tight_layout()
plt.show()
```

```python
#create a plot with a fig size 
fig_ure, axe_s = plt.subplots(2,1, figsize=(16,5))
for ax in axe_s:
    ax.plot(x,y)
```

```python
#how to save a figuee
fig_ure.savefig("enterName.extension")
```

```python
#adding Plot apperance
figuree = plt.figure()
axe = figuree.add_axes([0,0,1,1])
axe.plot(x,y, color ="r", linewidth = 2, linestyle = "-.", marker = "o" )
#set the x lim to the grapg 
axe.set_xlim([0,3])
```
