```python
#seperate out the data for a certain range for both the index as well as the  stock data 
idx = mcdon.loc["2001-1-1": "2002-1-1"].index
```

```python
stock = mcdon.loc["2001-1-1": "2002-1-1"]["Adj. Close"]
```

```python
#wew ill plot this using subpolts
fig, axe = plt.subplots(figsize = (19,5))
axe.plot_date(idx,stock, "-")
axe.xaxis.set_major_locator(dates.MonthLocator())
axe.xaxis.set_major_formatter(dates.DateFormatter("%b-%y"))
fig.autofmt_xdate()
```
