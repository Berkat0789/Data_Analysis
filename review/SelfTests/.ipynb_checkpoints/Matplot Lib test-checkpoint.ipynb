```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#create arraya
x = np.linspace(1,25,10)
y = x **2
```

```python
x
```

```python
y
```

```python
#linspace produces an evenly distributed range of numbers
```

```python
#functional method: 
plt.figure(figsize=(16,8))
plt.plot(x,y)
plt.xlabel("Xlabel")
plt.ylabel("Ylabel")
plt.title("Title for the grapg\n Here ")
plt.show()
```

```python
#creating multi plot 
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Graph 1")
plt.subplot(1,2,2)
plt.plot(y,x)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Graph 2")
plt.show()
```

```python
#create plot using axes 
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.3,0.5,0.5])
axes1.plot(x,y)
axes2.plot(y,x)
plt.show()
```

```python
#plot using object oriented programming 
figg, axx = plt.subplots(2,1)
axx[0].plot(x,y)
axx[0].set_xlabel("Xlabel")
axx[0].set_ylabel("Y_label")
axx[0].set_title("TiTle")
axx[1].set_xlabel("x_label")
axx[1].set_ylabel("Y_label")
axx[1].set_title("Title")
axx[1].plot(y,x)
plt.tight_layout()
```
