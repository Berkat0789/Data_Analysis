```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("resources/walmart_stock.csv", index_col= "Date", parse_dates=True)
```

```python
df
```

```python
df["Open"].plot(figsize =(16,6))
df.rolling(90).mean()["Close"].plot(figsize = (16,6))
```

```python
#ets discuss expanding 
df["Close"].expanding().mean().plot(figsize = (16,6))
```

```python
#create a bolunger band 
df['Close: 20day Mean'] = df["Close"].rolling(20).mean()
#Close 20 day moving average 
#upper = 20MA + 2*STD
df["Upper"] = df["Close: 20day Mean"]+ 2 * (df["Close"].rolling(20).std())
#lower = 20MA - 2*SDTD
df["Lower"] = df["Close: 20day Mean"] - 2 * (df["Close"].rolling(20).std())

#CLose 
df[["Close","Upper", "Lower", "Close: 20day Mean"]].tail(50).plot(figsize = (16,6))
```
